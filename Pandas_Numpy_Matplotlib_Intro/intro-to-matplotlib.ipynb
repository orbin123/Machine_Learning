# Intro to Matplotlib

### Data Visualization
Human brain can process data or information easily when it is in pictorial or graphical form.   

It helps us to quickly interpret the data and find the key trends.

### Matplotlib in Python
It is used plot data in 2d grphics.

Types:
* Bargraph 
* Histograms
* Scatter Plot
* Pie Plot 
* Hexagonal Bin Plot
* Area Plot

**Reference:** https://matplotlib.org/stable/tutorials/pyplot.html#sphx-glr-tutorials-pyplot-py


```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
arr1= np.array([1,2,3])
arr2= np.array([1,4,9])

# Plotting a line graph
plt.plot(arr1, arr2)

# Showing the plotted graph
plt.show()
```

```python
plt.plot(arr1, arr2, "--gD", markersize=10)
# plt.axis((1, 6, 0, 10))
plt.title("Square of numbers")
plt.ylabel("Squared Values")
plt.xlabel("Numbers")

plt.show()
```

#### Plotting Categorical Values


```python
names = ["Eminem", "Ed-Sheeran", "Weeknd"]
height = [162, 186, 175]

plt.figure(figsize=(9,3)) #figsize size of row and column in inch

plt.subplot(1,3,1) #(1-> no.of rows, 2-> no.of cols, 3->index)
plt.bar(names, height)
plt.subplot(1,3,2) #(1-> no.of rows, 2-> no.of cols, 3->index)
plt.scatter(names, height)
plt.subplot(1,3,3) #(1-> no.of rows, 2-> no.of cols, 3->index)
plt.plot(names, height)
plt.suptitle("Categorical Plotting")
plt.show()
```

### Axes, Legend and Grid

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,49,48,43,47]
min_t = [35,37,42,38,41,36,40]
avg_t = [42,44,47,44,45,39,43]
```

```python
plt.title("Weather")
plt.ylabel("Temperature")
plt.xlabel("Days")

plt.plot(days, max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Avg")

plt.legend()
plt.grid()
plt.show()
```

### Bar Chart

```python
company = ["Google", "Amazon", "Microsoft", "Meta"]
revenue = [190, 130, 180, 165]
profit = [50, 30, 60, 35]
xpos = np.arange(len(company))

xpos
```

```python
plt.title("US Tech Stocks")
plt.ylabel("Revenue(Bln)",)
plt.xticks(xpos, company)

plt.bar(xpos-0.2, revenue,width=0.4, label="Revenue")
plt.bar(xpos+0.2, profit,width=0.4 ,label="Profit")


plt.show()
```

### Histogram

```python
x = np.random.normal(170, 10, 250)
plt.hist(x)
plt.show() 
```

### Pie Charts

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels=mylabels, startangle = 90)
plt.show()
```
