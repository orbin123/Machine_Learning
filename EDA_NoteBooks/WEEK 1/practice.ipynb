```python
def minimumSum(num: int) -> int:
        str_num = str(num)
        char_array = list(str_num)  
        char_array.sort()  
        sorted_text = ''.join(char_array) 
        first = ''
        second = ''
        for i in range(len(sorted_text)):
                if i == 0 or i == 3:
                    first += sorted_text[i]
                else:
                    second += sorted_text[i]
        return int(first) + int(second)
                    
                
        
minimumSum(2932)
```
