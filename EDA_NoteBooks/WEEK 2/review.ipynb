```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

iris = sns.load_dataset('iris')

iris.head()
```

```python

sns.violinplot(x='petal_length', y ="petal_width" , data = iris.head(50))
```

```python
sns.boxplot(data=iris)
```

```python
sns.histplot(data=iris, x='sepal_length')
```

```python
# distribution skewness and kurtosis correlation and covariance
# eda
from scipy import stats
sepal_width	= iris['sepal_width']
skewness = stats.skew(sepal_width)
kurtosis = stats.kurtosis(sepal_width)

print(f'skewness: {skewness} and kurtosis: {kurtosis}')
```

```python
# correlation and covariance

corr_iris = iris[["sepal_length", "sepal_width",	"petal_length","petal_width"]].corr()

cov_iris = iris[["sepal_length", "sepal_width",	"petal_length","petal_width"]].cov()

sns.heatmap(data=corr_iris, annot=True)
```

```python
sns.heatmap(data=cov_iris, annot=True)
```

```python
# mean median mode

df = iris[["sepal_length", "sepal_width","petal_length","petal_width"]]

features = ["sepal_length", "sepal_width",	"petal_length","petal_width"]

def summary_analysis(df):
    summary ={}

    for col in features:
        summary[col] = {
            'mean': np.mean(df[col]),
            'Median': np.median(df[col]),
            'Mode': stats.mode(df[col])[0]
        }
    return pd.DataFrame(summary)

summary_analysis(df)
```
