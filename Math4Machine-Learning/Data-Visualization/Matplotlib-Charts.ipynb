### Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

data =  np.random.normal(loc=70, scale= 10 , size=1000)

plt.hist(data, bins=20, color="skyblue",edgecolor="black", )
plt.xlabel("Test Scores")
plt.ylabel("Frequency")
plt.title("Test score Frequency")
plt.show()
```

```python
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest

# Example dataset
X = [[1, 2, 3], [2, 3, 4], [3, 4, 5], [4, 5, 6]]
y = [0, 1, 0, 1]  # Categorical target variable

# Apply Chi-Square test
chi_scores, p_values = chi2(X, y)

for i, score in enumerate(chi_scores):
    print(f"Feature {i}: Chi-Square Score = {score}, p-value = {p_values[i]}")
```
