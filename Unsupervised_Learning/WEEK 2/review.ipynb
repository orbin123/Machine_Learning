```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA, TruncatedSVD
from sklearn.manifold import TSNE
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

X, y = load_iris(return_X_y=True)

X_scaled = StandardScaler().fit_transform(X)

pca_X = PCA(n_components=2).fit_transform(X)
tsne_X = TSNE(n_components=2, perplexity=20).fit_transform(X)
svd_X = TruncatedSVD(n_components=2).fit_transform(X)



pca_kmeans = KMeans(n_clusters=3)
pca_labels = pca_kmeans.fit_predict(pca_X)

tsne_kmeans = KMeans(n_clusters=3)
tsne_labels = pca_kmeans.fit_predict(tsne_X)

svd_label_kmeans = KMeans(n_clusters=3)
svd_label = pca_kmeans.fit_predict(svd_X)

plt.scatter(pca_X[:, 0], pca_X[:, 1], c=pca_labels, cmap='viridis')
```

```python
plt.scatter(tsne_X[:, 0], tsne_X[:, 1], c=tsne_labels, cmap='viridis')
```

```python
plt.scatter(svd_X[:, 0], svd_X[:, 1], c=svd_label, cmap='viridis')
```

```python
pca_kmeans_score = silhouette_score(pca_X, pca_labels)
pca_kmeans_score
```

```python
tsne_kmeans_score = silhouette_score(tsne_X, tsne_labels)
tsne_kmeans_score
```

```python
svd_kmeans_score = silhouette_score(svd_X, svd_label)
svd_kmeans_score
```
